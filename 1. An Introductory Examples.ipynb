# Quantecon - An Introductory Examples

Suppose we want to plot a white noise process $\epsilon_0, \epsilon_1, ... , \epsilon_T$, where each draw $\epsilon_t$ is independent standard normal.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.random.randn(100)
plt.plot(x)
plt.show()
```

Alternative version:
- The version below are less efficient but helps us to understand basic constructs like loops and illustrate common data types like lists.

***

## A Version with a For Loop

```python
ts_length = 100
epsilon_values = [] # Empty list

for i in range(ts_length):
    e = np.random.randn()
    epsilon_values.append(e)
    
plt.plot(epsilon_values)
plt.show()
```

## A Version with While Loop

```python
ts_length - 100
epsilon_values = []
i = 0
while i < ts_length:
    e = np.random.randn()
    epsilon_values.append(e)
    i = i + 1 # can be replaced by i += 1
plt.plot(epsilon_values)
plt.show()
```

## User-Defined Functions

```python
def generate_data(n):
    epsilon_values = []
    for i in range(n):
        e = np.random.randn()
        epsilon_values.append(e)
    return epsilon_values

data = generate_data(100)
plt.plot(data)
plt.show()
```

***

## Conditions

```python
def generate_data(n, generator_type):
    epsilon_values = []
    for i in range(n):
        if generator_type == 'U':
            e = np.random.uniform(0, 1)
        else:
            e = np.random.randn()
        epsilon_values.append(e)
    return epsilon_values

data = generate_data(100, 'U')
plt.plot(data)
plt.show()
```

### We can simplify the code:

```python
def generate_data(n, generator_type):
    epsilon_values = []
    for i in range(n):
        e = generator_type()
        epsilon_values.append(e)
    return epsilon_values

data = generate_data(100, np.random.uniform)
plt.plot(data)
plt.show()
```
